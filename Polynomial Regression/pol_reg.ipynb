# Polunomial Regression

# 1. Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# 2. Importing the Dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

# 3. Training Linear Regression on Dataset

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
```

# 4. Training Polynomial Regression on Dataset

```python
from sklearn.preprocessing import PolynomialFeatures
pr = PolynomialFeatures(degree=5)  # pow(x,2)
# connecting the PR with LR
xp = pr.fit_transform(x)  #making a new matrix
lr2 =  LinearRegression()
lr2.fit(xp,y)
```

# 5. Visualising PR result

```python
plt.scatter(x,y,color='red')
plt.plot(x,lr.predict(x),color='blue')
# Title and Lables
plt.title('Pos vs Sal (T|B) (LR)')
plt.xlabel('Position')
plt.ylabel('Salary')
# Display
plt.show()
```

# 6. Visualising PR result

```python
plt.scatter(x,y,color='red')
plt.plot(x,lr2.predict(xp),color='blue')
# Title and Lables
plt.title('Pos vs Sal (T|B)(PR)')
plt.xlabel('Position')
plt.ylabel('Salary')
# Display
plt.show()
```

# 7. Visualising PR result (higher resolution)

```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape((len(x_grid)),1)

plt.scatter(x,y,color='red')
plt.plot(x,lr2.predict(xp),color='blue')
# Title and Lables
plt.title('Pos vs Sal (T|B)(PR)')
plt.xlabel('Position')
plt.ylabel('Salary')
# Display
plt.show()
```

# 8. Predicting new result using LR 

```python
lr.predict([[6.5]])
```

# 9. Predicting new result using PR

```python
lr2.predict(pr.fit_transform([[6.5]]))
```
